```python
# Biểu diễn chiều dài, chiều rộng đài hoa trên không gian 2 chiều (chiều dài là hoành
# độ, chiều rộng là tung độ)

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("Iris.csv")

colors_dict = {'Iris-setosa': 'red', 'Iris-versicolor': 'blue', 'Iris-virginica': 'green'}

for specie, color in colors_dict.items():
    sub_df = df[df['Species'] == specie]
    plt.scatter(sub_df['SepalLengthCm'], sub_df['SepalWidthCm'], label=specie, color=color)

plt.title('Biểu đồ scatter của chiều dài và chiều rộng đài hoa')
plt.xlabel('Chiều dài đài hoa (cm)')
plt.ylabel('Chiều rộng đài hoa (cm)')
plt.legend(title='Loài hoa')

plt.show()
```

```python
# Bài 1a: Áp dụng với các bạn chưa/ít kinh nghiệm về lập trình
# Tính trung bình cộng chiều dài, chiều rộng đài hoa của từng loại (3 loại hoa)

means_sepal = df.groupby('Species')[['SepalLengthCm', 'SepalWidthCm']].mean()
print(means_sepal)
```

```python
# Bài 2a: Áp dụng với các bạn chưa/ít kinh nghiệm về lập trình
# Xác định chiều dài, chiều rộng đài hoa lớn nhất của từng loại (3 loại hoa)

maxs_sepal = df.groupby('Species')[['SepalLengthCm', 'SepalWidthCm']].max()
print(maxs_sepal)
```

```python
# Bài 3a: Áp dụng với các bạn chưa/ít kinh nghiệm về lập trình
# Nhập input đầu vào là 2 thông số, chiều dài & chiều rộng của 1 bông hoa bất kỳ (theo
# đơn vị mm). Xác định xem bông hoa đấy thuộc loại hoa nào (Iris setosa, Iris
# virginica hay Iris versicolor). Loại của bông hoa được xác định là class của

# a) Điểm có tọa độ gần nó nhất

# input_length = float(input('Please input length: '))
# input_width = float(input('Please input width: '))
input_length = 5.5
input_width = 2.5

distances = []
for index, row in df.iterrows():
    distance = np.sqrt(pow(row['SepalLengthCm'] - input_length, 2) + pow(row['SepalWidthCm'] - input_width, 2))
    distances.append((distance, row['Species']))

specie_of_input = min(distances)[1]
print(specie_of_input)
```

```python
# b) Trung bình cộng nhỏ nhất đến tất cả các điểm của từng loại

# input_length = float(input('Please input length: '))
# input_width = float(input('Please input width: '))
input_length = 5.5
input_width = 2.5

distances = []
for specie in df['Species'].unique():
    sub_df = df[df['Species'] == specie]
    distance = np.mean(np.sqrt(pow(sub_df['SepalLengthCm'] - input_length, 2) + pow(sub_df['SepalWidthCm'] - input_width, 2)))
    distances.append((distance, row['Species']))

specie_of_input = min(distances)[1]
print(specie_of_input)
```

```python
# Bài 1b: Áp dụng cho các bạn đã có kinh nghiệm về lập trình
# In ra chiều dài đài hoa lớn nhất của 10 bông hoa mỗi loại

def max_sepal_length(group):
    return group.nlargest(n=10, columns='SepalLengthCm')
    
print(df.groupby('Species').apply(max_sepal_length))
```

```python
# Bài 2b: Áp dụng cho các bạn đã có kinh nghiệm về lập trình
# Thống kê trong 50 bông hoa có tổng chiều dài + chiều rộng đài hoa lớn nhất, bao
# nhiêu bông thuộc về từng loại

df['Sum_Sepal_LenghtWidth'] = df['SepalLengthCm'] + df['SepalWidthCm']
top_50_species = df.nlargest(50, 'Sum_Sepal_LenghtWidth')['Species'].value_counts()
print(top_50_species)  
```
