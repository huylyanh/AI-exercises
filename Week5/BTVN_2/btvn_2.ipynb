```python
# Task 1: In ra index của 10 học sinh có math score và reading score đều nằm trong Top 100
import pandas as pd

df = pd.read_csv("dataset_2.csv")

top_100_math = df.sort_values(by="math score", ascending=False)
top_100_reading = df.sort_values(by="reading score", ascending=False)
top_10_both = pd.merge(top_100_math, top_100_reading, how='inner').head(10)
print(top_10_both)
```

```python
# Task 2: In ra index của 20 học sinh có điểm writing score cao nhất của mỗi group (race/ethnicity)
top_20 = lambda group: group.sort_values(by="writing score", ascending=False).head(20)

top_20_writing_by_race = df.groupby('race/ethnicity').apply(top_20)
print(top_20_writing_by_race)
```

```python
# Task 3: Thống kê số lượng học sinh của group A theo parental level of education
group_A = df[df['race/ethnicity'] == 'group A']
education_counts = group_A['parental level of education'].value_counts()
education_counts
```

```python
# Task 4: Có phải xu hướng chung, cha mẹ có học vấn càng cao thì điểm số trung bình 3 môn của con cái cũng càng cao ko

df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
edu_avg_score = df.groupby('parental level of education')['average score'].mean()
edu_avg_score
```

```python
# Task 5: Chất lượng bữa ăn (lunch) có tương phản với điểm số trung bình 3 môn của học sinh không
avg_score_by_lunch = df.groupby('lunch')['average score'].mean()
avg_score_by_lunch
```

```python
# Task 6: Tìm ra Top 10 học sinh có điểm toán cao nhất của mỗi group
top_10 = lambda group: group.sort_values(by="math score", ascending=False).head(10)

top_10_math_by_race = df.groupby('race/ethnicity').apply(top_10)
print(top_10_math_by_race)
```

```python
# Task 7: Với những học sinh ở cùng 1 group, cùng parental level of education, cùng loại bữa ăn (lunch), việc tham gia test preparation course có giúp học sinh đó có điểm trung bình 3 môn cao hơn những học sinh không tham gia không
def compare_test_pre(group, education, lunch):
    df_sub = df[(df['race/ethnicity'] == group) & (df['parental level of education'] == education) 
                & (df['lunch'] == lunch)]
    
    return df_sub.groupby('test preparation course')['average score'].mean()

groups = df['race/ethnicity'].unique()
educations = df['parental level of education'].unique()
lunchs = df['lunch'].unique()

for group in groups:
    for edu in educations:
        for lunch in lunchs:
            print('Group: {}. Education: {}. Lunch: {}.'.format(group, edu, lunch))
            print(compare_test_pre(group, edu, lunch))
            print('-----')
```
