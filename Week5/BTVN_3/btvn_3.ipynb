```python
# Task 1: Liệt kê các sản phẩm có doanh thu lớn hơn 100.000 Euro
import pandas as pd

df = pd.read_csv("dataset_3.csv")

df['revenue'] = df['price'] * df['units_sold']
revenue_greater_100K = df[df['revenue'] > 100000]
print(revenue_greater_100K[['title','revenue']])
```

```python
# Task 2: Liệt kê các sản phẩm có units_sold nằm trong top 100, đồng thời nằm trong nhóm 10% sản phẩm có rating cao nhất

top_100_unit_sold = df.sort_values(by="units_sold", ascending=False).head(100)
top_10_rating = df[df['rating'] > df['rating'].quantile(0.9)]
result = pd.merge(top_100_unit_sold, top_10_rating, how='inner')[['title', 'units_sold', 'rating']]
result
```

```python
# Task 3: Có phải sản phẩm càng được rating nhiều thì càng bán được nhiều sản phẩm không
correl_rating_units_sold = df['rating_count'].corr(df['units_sold'])
correl_rating_units_sold
```

```python
# Task 4: Thống kê Top 50 sản phẩm nằm trong rating từ 3.0 → 4.0 bán chạy nhất
df_filtered = df[(df['rating'] >= 3.0) & (df['rating'] <= 4.0)]
top_50 = df_filtered.sort_values(by='units_sold', ascending=False).head(50)[['title', 'rating', 'units_sold']]
top_50
```

```python
# Task 5: Có bao nhiêu sản phẩm có giá bán lớn hơn 50 EURO
result = df[df['price'] > 50]
print(result.shape[0])
```
