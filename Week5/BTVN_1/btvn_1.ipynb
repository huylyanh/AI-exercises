```python
# Task 1: Sắp xếp và in ra 10 thành phố có dân số lớn nhất & 10 thành phố có dân số nhỏ nhất
import pandas as pd

df = pd.read_csv("dataset_1.csv")

df['Population'] = df['Population'].replace(',', '', regex=True).astype(int)

top_10_biggest_population = df.sort_values(by='Population', ascending=False).head(10)[['City', 'Population']]
top_10_smallest_population = df.sort_values(by='Population', ascending=True).head(10)[['City', 'Population']]

print(top_10_biggest_population)
print("-----")
print(top_10_smallest_population)
```

```python
# Task 2: In ra tên các quốc gia có tối thiểu 3 thành phố trong danh sách này

country_counts = df['Country'].value_counts()
# print(country_counts)
country_with_3city = country_counts[country_counts.values > 3].index
country_with_3city = [city.strip() for city in country_with_3city]
country_with_3city
```

```python
# Task 3: In ra Top 5 quốc gia có nhiều thành phố xuất hiện trong bảng này nhất
top_5_countries = country_counts.nlargest(n=5)
top_5_countries
```

```python
# Task 4: In ra các thành phố có dân số & diện tích đều nằm trong Top 20
df['Area   M2'] = pd.to_numeric(df['Area   M2'], errors='coerce')
df['Area   M2'].fillna(df['Area   M2'].median())

top_20_population = df.nlargest(20, 'Population')[['City', 'Population']]
top_20_area = df.nlargest(20, 'Area   M2')[['City', 'Area   M2']]
top_20_both = pd.merge(top_20_population, top_20_area, on='City')
print(top_20_both)
```

```python
# Task 5: Thống kê mật độ dân số theo quốc gia
df['Density KM2'] = df['Density KM2'].replace(',', '', regex=True)
df['Density KM2'] = df['Density KM2'].astype(float)
density_by_country = df.groupby('Country')['Density KM2z'].mean()
density_by_country
```

```python
# Task 6: Thống kê các thành phố có dân số lớn nhất của từng quốc gia (chỉ tính của những quốc gia có 2
# thành phố xuất hiện trở lên trong bảng)
df['Population'] = df['Population'].replace(',', '', regex=True)
df['Population'] = df['Population'].astype(int)

country_counts = df['Country'].value_counts()
country_with_multiple_city = country_counts[country_counts > 1].index
filtered_df = df[df['Country'].isin(country_with_multiple_city)]

largtest_city = filtered_df.loc[filtered_df.groupby('Country')['Population'].idxmax()]
largtest_city
```
